```python
import os

!pip install -U openai-whisper
import whisper

from google.colab import drive
drive.mount('/content/drive/')
```

```python
model = whisper.load_model("large")                 # ("tiny") ("base") ("small")  ("medium") ("large")
os.chdir('/content/drive/MyDrive/Audio/Датасет_аудио (1)')    # Установить текущий каталог слеш прямой /
```

```python
# Определить папки акустических файлов и для распознанного текста
audio_dir = "/content/drive/MyDrive/Audio/Датасет_аудио (1)"  # Папка с исходными аудио файлами
text_dir = "/content/drive/MyDrive/Audio/Датасет_текст"   # Папка для текстовых файлов

# Создать папку для текстовых файлов, если она не существует
if not os.path.exists(text_dir):
    os.makedirs(text_dir)

# Получить список всех акустических файлов в папке "Датасет_аудио"
audio_files = [f for f in os.listdir(audio_dir) if f.endswith(".ogg")]

# Транскрибировать каждый исходный акустический файл
for audio_file in audio_files:
    transcription = model.transcribe(audio_file)   # Провести  транскрибацию

# Print the result
    print(f"File: {audio_file}")
    print(transcription["text"])           # Напечатать текст аудио запроса
    print("="*50)
# На транскрибацию всех 500 файлов ушло 31 минуты 2 секунды
```

```python
# Функция определения правильно распознанного текста в процентах
def calculate_accuracy(original_text, recognized_text):
    original_words = original_text.split()       # Оригинальный текст
    recognized_words = recognized_text.split()   # Распознанный текст

    # Минимальная длина текста (оригинального или распознанного)
    min_length = min(len(original_words), len(recognized_words), 25)

    # Счетчик корректно распознанных слов в тексте
    correct_count = sum(1 for word1, word2 in zip(original_words[:min_length], recognized_words[:min_length]) if word1 == word2)

    accuracy = (correct_count / min_length) * 100  # Процент правильно распознанных слов

    return accuracy                                # Вернуть процент правильно распознанных слов

original_text = "Это оригинальный текст"
recognized_text = "Это текст распознанный"

accuracy = calculate_accuracy(original_text, recognized_text)
print(f"Точность распознавания: {accuracy:.2f}%")
```
