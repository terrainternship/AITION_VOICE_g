```python
from pydub import AudioSegment
import os
import speech_recognition as sr

# Function to convert .ogg to .wav
def convert_ogg_to_wav(ogg_file, wav_file):
    sound = AudioSegment.from_ogg(ogg_file)
    sound.export(wav_file, format="wav")

# Function to perform speech recognition using Google Web Speech API
def recognize_speech(audio_file):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio_data = recognizer.record(source)
        try:
            text = recognizer.recognize_google(audio_data, language="ru")
            return text
        except sr.UnknownValueError:
            return "Speech Recognition could not understand audio"
        except sr.RequestError as e:
            return f"Could not request results from Google Speech Recognition service; {e}"

# Folder containing .ogg files
ogg_folder = '/content/drive/MyDrive/Audio/Датасет_аудио (1)'

# Folder to save .wav files
wav_folder = '/content/drive/MyDrive/Audio/Google_data_audio'

# Ensure the output folder exists
os.makedirs(wav_folder, exist_ok=True)

# Process each .ogg file in the folder
for ogg_file in os.listdir(ogg_folder):
    if ogg_file.endswith(".ogg"):
        ogg_path = os.path.join(ogg_folder, ogg_file)
        wav_file = os.path.join(wav_folder, os.path.splitext(ogg_file)[0] + ".wav")

        # Convert .ogg to .wav
        convert_ogg_to_wav(ogg_path, wav_file)

        # Perform speech recognition
        result = recognize_speech(wav_file)

        # Print the result
        print(f"File: {ogg_file}")
        print("Speech Recognition Result:", result)
        print("="*50)
```

```python
# Функция определения правильно распознанного текста в процентах
def calculate_accuracy(text, result):
    original_words = text.split()       # Оригинальный текст
    recognized_words = result.split()   # Распознанный текст

    # Минимальная длина текста (оригинального или распознанного)
    min_length = min(len(original_words), len(recognized_words), 25)

    # Счетчик корректно распознанных слов в тексте
    correct_count = sum(1 for word1, word2 in zip(original_words[:min_length], recognized_words[:min_length]) if word1 == word2)

    accuracy = (correct_count / min_length) * 100  # Процент правильно распознанных слов

    return accuracy                                # Вернуть процент правильно распознанных слов

text = "Это оригинальный текст"
result = "Это текст распознанный"

accuracy = calculate_accuracy(text, result)
print(f"Точность распознавания: {accuracy:.2f}%")
```
